```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resto.csv')
```

```python
print("Max votes:", df['Votes'].max())
print("Min votes:", df['Votes'].min())
```

```python
highest_votes = df[df['Votes'] == df['Votes'].max()]
print("Restaurant(s) with highest votes:")
print(highest_votes[['Restaurant Name', 'Votes', 'City', 'Cuisines']])
```

```python
lowest_votes = df[df['Votes'] == df['Votes'].min()]
print("Restaurant(s) with lowest votes:")
print(lowest_votes[['Restaurant Name', 'Votes', 'City', 'Cuisines']])
```

```python
top_votes = df.sort_values('Votes', ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(data=top_votes, x='Votes', y='Restaurant Name', palette='viridis')
plt.title('Top 10 Restaurants with Highest Votes')
plt.xlabel('Votes')
plt.ylabel('Restaurant Name')
plt.tight_layout()
plt.show()
```

```python
df = df[['Votes', 'Aggregate rating']].dropna()
```

```python
correlation = df['Votes'].corr(df['Aggregate rating'])
print(f"Correlation between Votes and Rating: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Votes', y='Aggregate rating', alpha=0.6)
plt.title(f'Votes vs Rating (Correlation = {correlation:.2f})')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```
