```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Resto.csv')
```

```python
grouped = df.groupby('Price range')[['Has Online delivery', 'Has Table booking']].mean().reset_index()
```

```python
print(grouped)
```

```python
plt.figure(figsize=(10,5))

# Online Delivery
plt.subplot(1, 2, 1)
sns.barplot(data=grouped, x='Price range', y='Has Online delivery', palette='Blues')
plt.title('Price Range vs Online Delivery')
plt.ylabel('Proportion with Online Delivery')

# Table Booking
plt.subplot(1, 2, 2)
sns.barplot(data=grouped, x='Price range', y='Has Table booking', palette='Greens')
plt.title('Price Range vs Table Booking')
plt.ylabel('Proportion with Table Booking')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Online Delivery
plt.subplot(1, 2, 1)
sns.barplot(data=grouped, x='Price range', y='Has Online delivery', palette='coolwarm')
plt.title('Online Delivery by Price Range')
plt.ylabel('Proportion Offering Online Delivery')

# Table Booking
plt.subplot(1, 2, 2)
sns.barplot(data=grouped, x='Price range', y='Has Table booking', palette='viridis')
plt.title('Table Booking by Price Range')
plt.ylabel('Proportion Offering Table Booking')

plt.tight_layout()
plt.show()
```
